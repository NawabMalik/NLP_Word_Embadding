```python
sentence=['The glass of water','The glass of orange','The cop of coffee','i am a handsome boy','i am a good listiener',
          'understanding the meaning of words','your thought is good']
sentence
```

```python
# lets convert the sentence into one hot encoding first:
from tensorflow.keras.preprocessing.text import one_hot
voc_size=10000
```

```python
one_hot=[one_hot(word,voc_size) for word in sentence]
one_hot
```

```python
# lets convert this sparse or high dimentional matrix into low dimentional or dense matrix using embedding layer
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences  
#using padding to control 4 or 5 values of one hot encoding result into equal values:
from tensorflow.keras.models import Sequential
```

```python
padding=pad_sequences(one_hot, maxlen=7, padding='pre')    # 0 added as prefix, can use padding='post' to add 0 as at last:
padding
```

```python
model=Sequential()
model.add(Embedding(voc_size, output_dim=10, input_length=7))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(padding)
```
